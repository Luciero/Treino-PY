# Mais edições de Gráfico com MatplotLib

```python
#importando o matplotlib
import matplotlib.pyplot as plt
```

- Usando o módulo numpy para trabalhar com números

```python
#importando o módulo numpy
import numpy as np
```

### Outros tipos de Gráfico:

- Linha

```python
vendas = np.random.randint(1000, 3000, 50)  #Gera uma lista aleatoria (Min, Max, Quantidade)
meses = np.arange(1, 51)  #gera uma lista de o inicio ate o fim (Min incluso, Max nao incluso)

print(vendas)
print()
print(meses)
```

```python
plt.plot(meses,vendas)
plt.axis([0,50,0,max(vendas)+ 500])
plt.xlabel('meses')
plt.ylabel('vendas')
plt.show()
```

- Editando o Gráfico de Linha

```python
plt.plot(meses,vendas,'r.-')
plt.axis([0,50,0,max(vendas)+ 500])
plt.xlabel('meses')
plt.ylabel('vendas')
plt.show()
```

```python
#editando o tamanho e a cor da linha - quais parâmetros posso editar?
```

- Dispersão

```python
plt.scatter(meses,vendas)
plt.show()
```

- Barras

```python
plt.bar(meses,vendas)
plt.show()
```

- Trabalhando com Múltiplos Gráficos no mesmo "Plot" -> Para melhor visualização/comparação

```python
plt.figure(1,figsize=(17,4.5))
plt.subplot(1,3,1) #exibe varios graficos (quantas linhas, quantas colunas, numero do grafico)
plt.plot(meses,vendas,'r.-')
plt.axis([0,50,0,max(vendas)+ 500])
plt.xlabel('meses')
plt.ylabel('vendas')
plt.subplot(1,3,2)
plt.scatter(meses,vendas)
plt.xlabel('meses')
plt.ylabel('vendas')
plt.subplot(1,3,3)
plt.bar(meses,vendas)
plt.xlabel('meses')
plt.ylabel('vendas')
plt.show()
```
