# Jogo da velha

- Definir o jogo (matriz)
- Definir X e O para colocar na matriz
- Colocar na matriz
- imprimir e verificar ganhador ou perdedor

```python
matriz = [[' ' for x in range(3)] for y in range(3)] 


def gameTurn(m,a,b,j):
    if j == 1:
        m[a-1][b-1] = 'X' 
        printGame(m)
        j += 1
        return j
    else:
        m[a-1][b-1] = 'O' 
        printGame(m)
        j -= 1
        return j
     
def printGame(m):
    print('\n')

    for i in range(3):
        control = 0
        
        for j in range(3):
            print("",m[i][j],end="")
            if control < 2:
                print('  | ',end="")
            control += 1
            
        if i < 2:    
            print("\n---------------")
    print('\n\n')

        
def verifyWin(m,c):
    for i in range(3):
        for j in range(3):
            if m[0][0] == 'X' and m[0][1] == 'X' and m[0][2] == 'X':
                return True
        
            elif m[0][0] == 'X' and m[1][1] == 'X' and m[2][2] == 'X':
                return True
        
            elif m[0][0] == 'X' and m[1][0] == 'X' and m[2][0] == 'X':
                return True
        
            elif m[0][1] == 'X' and m[1][1] == 'X' and m[2][1] == 'X':
                return True
        
            elif m[0][2] == 'X' and m[1][1] == 'X' and m[2][0] == 'X':
                return True
        
            elif m[1][0] == 'X' and m[1][1] == 'X' and m[1][2] == 'X':
                return True
        
            elif m[2][0] == 'X' and m[2][1] == 'X' and m[2][2] == 'X':
                return True
        
            elif m[0][2] == 'X' and m[1][2] == 'X' and m[2][2] == 'X':
                return True
        
        
            elif m[0][0] == 'O' and m[0][1] == 'O' and m[0][2] == 'O':
                return True
        
            elif m[0][0] == 'O' and m[1][1] == 'O' and m[2][2] == 'O':
                return True
        
            elif m[0][0] == 'O' and m[1][0] == 'O' and m[2][0] == 'O':
                return True
        
            elif m[0][1] == 'O' and m[1][1] == 'O' and m[2][1] == 'O':
                return True
        
            elif m[0][2] == 'O' and m[1][1] == 'O' and m[2][0] == 'O':
                return True
        
            elif m[1][0] == 'O' and m[1][1] == 'O' and m[1][2] == 'O':
                return True
        
            elif m[2][0] == 'O' and m[2][1] == 'O' and m[2][2] == 'O':
                return True
        
            elif m[0][2] == 'O' and m[1][2] == 'O' and m[2][2] == 'O':
                return True
                
            
            elif m[i][j] != ' ':
                c += 1
                if c == 9:
                    c = 0
                    return 'Draw'

                
def clearMatriz(m):
    for i in range(3):
        for j in range(3):
            m[i][j] = ' '

def main():
    jogador = 1
    control = 0
    
    while True:
        x,y = map(int,input().split())
        if (x <= 3 and x > 0) and (y <= 3 and y > 0):
            if matriz[x-1][y-1] == ' ':
                result = False
                jogador = gameTurn(matriz,x,y,jogador)
                result = verifyWin(matriz,control)
                
                if result == True and jogador == 1:
                    print('\nJogador nº {} ganhou!!'.format(jogador+1))
                    break
                elif result == True and jogador == 2:
                    print('\nJogador nº {} ganhou!!'.format(jogador-1))
                    break
                    
                elif result == 'Draw':
                    clearMatriz(matriz)
                    print('\nEmpate!! Comece novamente\n')
            else:
                print('\nErro, digite uma coordenada que ainda não foi usada!\n')
        else:
            print('\nErro digite um numero entre 1 e 3 para as coordenadas X e Y!\n')
main()
```
