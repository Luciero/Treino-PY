### Executar Comandos Javascript (Scroll da tela)

- Você consegue, por meio do Selenium, executar comandos javascript no seu navegador

- Isso é essencial para dar scroll na tela, por exemplo, caso seja necessário, como no youtube

```python
# criando o navegador
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time

servico = Service(ChromeDriverManager().install())
navegador = webdriver.Chrome(service=servico)

navegador.get('https://www.youtube.com')
time.sleep(1)

pesquisa = navegador.find_element(By.NAME, 'search_query')
pesquisa.send_keys('python')
pesquisa.send_keys(Keys.ENTER)
time.sleep(3)

# Dar um scroll na pagina
for i in range(10):
    scroll = 4000 * i
    navegador.execute_script(f'window.scroll(0, {scroll});')
    time.sleep(2.5)

videos_links = navegador.find_elements(By.ID, 'thumbnail')
for links in videos_links:
    if links.get_attribute('href') != None:
        print(links.get_attribute('href'))
```
